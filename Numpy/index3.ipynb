### Advance operation with Business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  Data Strucutre : [resturant_id,2021,2022,2023,2024]
sales_Data =np.array([
    [1,12000,33301,89562,74896],
    [2,13000,44301,99562,34896],
    [3,14000,55301,79562,14896],
    [4,15000,77301,69562,24896],
    [5,16000,66301,59562,44896]
])
print("=== Zomato Sales Analysis===")
print("\n Sales data shape",sales_Data.shape)
print("\n Sales data from 1st to 3rd\n",sales_Data[:, 1:])
print("\n Sales data from 1st to 3rd\n",sales_Data[:3])
```

```python
# totale sales  per year
year_total = np.sum(sales_Data[: ,1:], axis=0)
print("total sale in year",year_total)

# Minimum sale per restaurant
min_sales = np.min(sales_Data[: ,1:],axis=1)
print("minmum sale per restaurant",min_sales)

#Maximum sale per year
max_sale = np.max(sales_Data[: ,1:],axis=0)
print("maximum sale per yaer",max_sale)

# Average sales per restaurant
avg_sale = np.mean(sales_Data[: , 1:],axis=1)
print("average sale of restaurant",avg_sale)
```

```python
# cumulative sale
cumsum =np.cumsum(sales_Data[:,1:],axis=1)
print("cumulate sale",cumsum)

plt.Figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average  cumulative sale  accross all restaurant")
plt.xlabel("year")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

## Vector Addition

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("vector addition",vector1+vector2)

print("\nvector multiplication",vector1*vector2)

print("\nvector dot product",np.dot(vector1,vector2))

angle = np.arccos(np.dot(vector1,vector2) / (np.linalg.norm(vector1)* np.linalg.norm(vector2)))

print("angle",angle)
```
