# Linear Regression

```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
X = pd.read_csv('./Trainingdata/Linear_X_Train.csv').values.reshape((-1,))
Y = pd.read_csv('./Trainingdata/Linear_Y_Train.csv').values.reshape((-1,1))
```

```python
Y.shape
```

```python
plt.scatter(X, Y, color='red')
plt.show()
```

```python
w1 = 50 
w0 = 1
alpha =0.1
yp = w1 * X + w0
plt.scatter(X, Y,c='r')
plt.plot(X,yp,c='b')
plt.show()
```

```python
w1 =  w1 -alpha * np.mean((yp - Y) * X)
w0 =  w0 -alpha * np.mean(yp - Y)
print(w1,w0)
```

```python
df = pd.read_csv('./Trainingdata/BostonHousing.csv')
df.head(5)
```

```python
%matplotlib inline
plt.scatter(df.age,df.tax,c='r',marker='+')
plt.xlabel("age")
plt.ylabel("tax")
plt.show()
```

```python
reg  = LinearRegression() # obj of linear regression
reg.fit(df[['age']],df.tax) #fit  mean u are training the LR_model using the available data points
```

```python
reg.predict([[45]])
```

```python
%matplotlib inline
plt.scatter(df.age,df.tax,c='r',marker='+')
plt.plot(df.age,reg.predict(df[['age']]),c='b')
plt.xlabel("age")
plt.ylabel("tax")
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(df.tax, reg.predict(df[['age']])) # r2_score is used to check the accuracy of the model
```
