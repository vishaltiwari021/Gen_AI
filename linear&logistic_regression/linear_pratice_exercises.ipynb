## Pratice Exercises

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from word2number import w2n
```

```python
df = pd.read_csv('./Trainingdata/homeprices.csv')
df
```

```python
df.bedrooms.median()
```

```python
import math
median_bedroom = math.floor(df.bedrooms.median())
median_bedroom
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedroom)
df
```

```python
reg  = LinearRegression()
reg.fit(df[['area','bedrooms','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3350,6,35]])
```

## Exercises 2

```python
data  = pd.read_csv('./Trainingdata/hiring.csv')
data
```

```python
import math
test_Score_mean = math.floor(data.test_score.mean())
test_Score_mean 
```

```python
data.test_score = data.test_score.fillna(test_Score_mean)
data
```

```python
data.experience = data.experience.fillna("zero")
data
```

```python
data.experience = data.experience.apply(w2n.word_to_num)
data
```

```python
clf  = LinearRegression()
clf.fit(data[['experience','test_score','interview_score']],data['salary'])
```

```python
clf.predict([[10,11,5]])
```

## this is how to save your model

```python
import joblib
```

```python
joblib.dump(clf, 'model_joblib.pkl')
```

```python
md =  joblib.load('model_joblib.pkl')
```

```python
md.predict([[11,6,20]])
```

## Exercise 3

```python
dd = pd.read_csv('./Trainingdata/homeprices2.csv')
dd.head()
```

```python
dummies = pd.get_dummies(dd.town)
dummies
```

```python
merge  = pd.concat([dd,dummies],axis=1)
merge
```

```python
final= merge.drop(['town','west windsor'],axis=1)
final
```

```python
model = LinearRegression()
```

```python
X = final.drop(['price'],axis=1)
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[2656,0,0]])
```

```python
model.score(X,y)
```
