# Matplotlib
 *line plot

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.arange(1,10)
x
```

```python
y1= x*2
y2= x*3
y1,y2
```

```python
plt.plot(x,y1)
plt.title("line graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y1,color='r',linewidth=2,linestyle=':')
plt.title("line graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y1,color='r',linewidth=2,linestyle=':')
plt.plot(x,y2,color='g',linewidth=2,linestyle='-.')
plt.title("line graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='r',linewidth=2,linestyle=':')
plt.subplot(1,2,2)
plt.plot(x,y2,color='g',linewidth=2,linestyle='-.')
plt.title("line graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
X = np.linspace(-3,3,50)
y1 = 2*X+4
y2 = X**2 - 2

plt.plot(X,y1,label = 'y1')
plt.plot(X,y2,label = 'y2')

plt.grid()
plt.title('Some Functions')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```

# some other types in Gen AI class of graph

```python
x = np.arange(1,11)
y1 = [10,5,15,20,13,16,27,18,23,30]
y2 = 2*x+3
plt.plot(x,y1,'r:o',label = 'Price')
plt.plot(x,y2,linestyle = '--',color = 'orange',label = 'Line',marker = '^')

plt.xlabel('X')
plt.ylabel('f(X) = Y')
plt.title("Line Plot")
plt.legend()
# plt.grid()
plt.show()
```

```python
font = {'family': 'serif',
        'color':  'blue',
        'weight': 'normal',
        'size': 16,
        }
x = np.arange(1,11)
y1 = 3*x+5
y2 = 2*x+3
plt.plot(x,y1,label = 'Price',color = 'r',marker = '*',markersize = 15,markerfacecolor = 'b')
plt.plot(x,y2,label = 'Line',color = 'g',marker = 'o',linestyle=':')

plt.xlabel('X',fontdict=font)
plt.ylabel('f(X) = Y',fontdict = font)
plt.title("Line Plot",fontdict = font)
plt.legend()
# plt.grid()
plt.savefig('LinePlot.jpg')
plt.show()
```
