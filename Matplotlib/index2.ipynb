```python
import numpy as np
import matplotlib.pyplot as plt 
```

# Bar Plot

```python
student  = {'bob' :87,'lil':85,'cos':99}
```

```python
name = list(student.keys())
values = list(student.values())

name,values
```

```python
plt.bar(name,values)
plt.title("bar graph")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(name,values,color='r')
plt.title("bar graph")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

## Scatter plot

```python
x = np.arange(1,11)
y1 = [10,5,15,20,13,16,27,18,23,30]

plt.scatter(x,y1,c='yellow',marker='*',s=200)
plt.title("scatter graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x = np.arange(1,11)
y1 = [10,5,15,20,13,16,27,18,23,30]
y2 = [14,5,15,2,13,6,17,8,3,0]
plt.subplot(121)
plt.scatter(x,y1,c='yellow',marker='*',s=200)
plt.subplot(122)
plt.scatter(x,y2,c='g',marker='.',s=100)
plt.title("scatter graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

## Histogram


```python
data  = np.array([1,3,4,7,8,10,12,15,16,20])
data
```

```python
plt.hist(data)
plt.title("histogram")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

## BOX plot

```python
l1  = np.array([1,3,5,7,9])
l2  = np.array([0,2,4,6,8])
l3  = np.array([2,5,7,11,13])

arr = l1,l2,l3
arr
```

```python
plt.boxplot(arr)
plt.title("boxplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.violinplot(arr,showmedians=True)
plt.title("boxplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

## pie plot

```python
fruite = ["apple","banana","orange","grape"]
qunatity = [10,20,30,40]

plt.pie(qunatity,labels=fruite,autopct='%1.1f%%',startangle=140)

plt.show()
```

```python
# doughnut
plt.pie(qunatity,labels=fruite,autopct='%1.1f%%',startangle=140,shadow=True,radius=1)
plt.pie([1],colors=['w'],radius=0.5)
plt.show
```
